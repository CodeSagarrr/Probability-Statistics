# Probablity Basics

* Probability ka matlab hota hai kisi event (ghatna) ke hone ki sambhavna. Matlab koi kaam hone ke kitne chances hain.
* Value hamesha 0 se 1 ke beech hoti hai —
   * * 0 matlab event kabhi nahi hoga,
   * * 1 matlab event zarur hoga.

* Why we use in AI/ML:
AI/ML models real world ke uncertain data ke saath kaam karte hain. Probability help karti hai model ko prediction me confidence samajhne me.
Example: Model ne bola “ye email spam hone ki probability 0.9 hai” , matlab model 90% sure hai ki ye spam hai.


# Random Experiments

* Random experiment wo process hai jiska result uncertain ho hum predict nahi kar sakte pehle se kya hoga.
Har bar karne par result change ho sakta hai.
* * Coin toss karna
* * Weather prediction
* Why in AI/ML:
AI models bhi ek tarah ke random experiments jese behave karte hain — unka output uncertain hota hai aur probability se samjha ja sakta hai.

# Sample Space

* Sample space (S) wo set hai jisme sab possible outcomes hote hain random experiment ke.
* * Coin toss: S = {Head, Tail}
* * Dice roll: S = {1, 2, 3, 4, 5, 6}
* Why in AI/ML:
Sample space help karta hai hume define karne me ki model ke sab possible outcomes kya ho sakte hain.
Example: Model “cat” ya “dog” predict karega, to sample space = {cat, dog}

# Events

* Event matlab sample space ka ek specific subset , matlab particular outcome ya outcome group jisme hume interest hai.
* * Dice roll me event “even number aana”: E = {2, 4, 6}
* * Coin toss me event “Head aana”: E = {Head}
* Why in AI/ML:
AI me event ek result ya condition hota hai jisko model predict karta hai.
Example: Email spam model me “Email is spam” ek event hai.

# Example

```python
import random
# we are taking dice example 
dice_number = random.randint(1 , 6)
print("Dice number : " , dice_number)

# Sample space 
S = [ 1, 2, 3, 4, 5, 6 ]
print("Smaple Space :" , S)

# Events : Getting even number
event_even = [ 2, 4, 6 ]
print("Event  :" , event_even)

# Calculate probablity 
probablity = len(event_even) / len(S)
print(f"Probablity of even number " , probablity)
```
