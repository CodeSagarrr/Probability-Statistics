# Random Variables

* Random Variable ek aisi variable hoti hai jiska value kisi random experiment ke outcome pe depend karta hai.

* Matlab jab hum experiment karte hain, to result ke hisab se variable ek numeric value leta hai.


## Types

* 1 : Discrete Random Variable – jisme limited ya countable values hoti hain.

    - Example: Dice roll → values {1, 2, 3, 4, 5, 6}

* 2 : Continuous Random Variable – jisme infinite possible values hoti hain ek range ke andar.

    - Example: Temperature, height, time, etc.

* Why in AI/ML:

    - AI me random variables use hote hain uncertain outcomes ko represent karne ke liye. 
    
* Example:

    - Spam probability = 0.8 → Random variable ke form me likhte hain.

    - Model ke output (0 to 1 probability) ek random variable ke tarah behave karte hain.

* Simple Example:

    - Coin toss:

        - Random variable X = 1 agar Head aaya, 0 agar Tail aaya.
        
        - Yaha X ek discrete random variable hai.

```python
import random

def count_choice():
    x = random.choice([0,1])
    return x

result = [ count_choice() for _ in range(10) ]
print("Coin toss results (Random Variable X):" , result)
```

# Probablity Distribution

* Probability distribution batata hai ki ek random variable ke har possible value ke hone ki probability kya hai.

* Ye basically “pattern” batata hai random variable ke behavior ka.

* Why in AI/ML:

     - ML models me distribution help karta hai:

     - Data ka spread samajhne me

     - Model ke assumption verify karne me (jaise Normal Distribution)

     - Probabilistic models build karne me (Bayesian Networks, Gaussian Mixture Models, etc.)

## 3. Important Distributions

# Bernoulli Distribution

* Ye distribution tab hoti hai jab sirf 2 possible outcomes hote hain — success (1) ya failure (0).

* AI/ML Use:

    - Binary classification me use hoti hai.
    
    - Example: Logistic Regression output 0 ya 1 ke form me hota hai.

```python
# Example of Bernoulli Distribution

d1 = [ random.choice([0,1]) for _ in range(10) ]
print("Bernoulli sample:", d1)
```

# Binomial Distribution

* Multiple Bernoulli trials ka sum. Matlab same experiment ko repeat karna aur count karna kitni baar success hua.

* Example:

    - 10 coin toss me kitni baar Head aaya

* AI/ML Use:

    - Feature sampling aur binary outcome problems me.

```python
from numpy.random import binomial

# n = 10 tosses (no. of tosses), p = 0.5 (probability of head)
result = binomial(n = 10 , p = 0.5 , size=10)
print("Binomial sample :" , result)
```

# Normal (Gaussian) Distribution

* Ye sabse famous distribution hai — bell-shaped curve jise “Normal Curve” bolte hain.

* Most real-world data (height, weight, exam marks) normal distribution follow karte hain.

* Why in AI/ML:

    - Many algorithms assume data is normally distributed

    - Used in Gaussian Naive Bayes, Linear Regression, etc.

* Example:

    - Student marks — most students average marks lete hain, thode kam, thode zyada.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=50 , scale=10 , size=1000)
plt.hist(data , bins=30)
plt.title("Normal Distribution")
plt.show()
```

# Uniform Distribution

* Sab outcomes ke equal chance hone par.

    - Example: Dice roll → har number ke equal chance (1/6)

* AI/ML Use:

    - Random initialization of weights, random sampling, Monte Carlo methods me use hoti hai.

```python

data = np.random.uniform(0 , 1 , size=5)
print("Uniform sample:", data)
```

# Poisson Distribution

* Ye distribution tab use hoti hai jab hum count karte hain ki ek fixed time me kitni baar koi event hua.
* Example:

    - 1 minute me kitne calls aaye
    
    - Website per ek second me kitne log aaye

* AI/ML Use:

    - Event prediction models, anomaly detection me use hoti hai.

```python
x = np.random.poisson(lam=2 , size=5)
print("Poisson sample:", x)
```
