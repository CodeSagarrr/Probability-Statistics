# Joint and Conditional distribution

## Joint distribution

* Joint distribution batata hai ki do ya zyada random variables ek saath kaise behave karte hain.
Matlab ek variable ki value dusre variable ke sath kis tarah related hai.

* Example:

    - X = Student’s study hours

    - Y = Student’s marks

    - Joint distribution batata hai ki kis study hour pe kitne marks aate hain.

* Why in AI/ML:

    - Joint probability se models variables ke relation samajhte hain — jaise feature aur label ke beech relation.
    
    - Example: Naive Bayes classifier P(X, Y) ke concept pe hi based hai.


```python
import numpy as np
import seaborn as sns
import pandas as pd

# Two random variables
x = np.random.normal(50 , 10 , 1000)
y = x + np.random.normal(0 , 5, 1000)

pf = pd.DataFrame({ "X" : x , "Y" : y})
sns.jointplot(data=pf , x="X" , y="Y" , kind="scatter")
```

## Conditional Distribution

* Conditional probability batata hai ki ek variable ki probability kya hai jab dusra variable fixed ho.

* Formula:

    - P(A | B) = P(A and B) / P(B)

* Example:

    - P(Student gets A grade | studies 5 hours daily)

* Why in AI/ML:

    - AI models mostly conditional probabilities pe based hote hain — jaise model ke prediction ka chance given data.

    - Example: P(Spam | Words in email)

```python
# Simple probability example
p_spam = 0.3
p_word_given_spam = 0.8
p_word = 0.5

p_spam_given_word = (p_word_given_spam * p_spam) / p_word
print("P(Spam | Word):", p_spam_given_word)
```

# Sampling and Estimation 

## Sampling

* Sampling matlab population ke ek chhote hissa (subset) ko lena, jisse poori population ke baare me estimate kiya ja sake.

* Example:

    - 1000 students ke marks me se 100 ka sample lena to estimate average marks.

* Why in AI/ML:

    - Model training me hum poore real-world data nahi le sakte — sampling se efficient training hoti hai.

## Estimation

* Estimation ka matlab hota hai sample ke basis pe population ke parameter (mean, variance, etc.) ko guess karna.

* Types:

    - Point Estimation: Single value estimate (mean = 70)

    - Interval Estimation: Range estimate (mean between 68–72)

```python
data = np.random.normal(50 , 10 , 1000)
sample = np.random.choice(data , 100)
estimate_mean = np.mean(sample)

print("Estimated Mean:", estimate_mean)
```

## Hypothesis Testing

* Hypothesis testing ek statistical method hai jisse hum check karte hain ki koi assumption true hai ya nahi.

* Steps:

    - Null Hypothesis (H₀): koi change nahi hai.

    - Alternative Hypothesis (H₁): kuch change ya effect hai.

    - Test statistic compute karte hain.

    - P-value check karte hain (agar p < 0.05 to reject H₀).

* Example:

    - Check karna ki new model old model se better perform karta hai ya nahi.

```python
from scipy import stats

old_model = np.random.normal(50 , 5 , 100)
new_model = np.random.normal(72 , 5 , 100)

t_stat , p_value = stats.ttest_ind(old_model , new_model)
print("p-value:", p_value)
```

## Statistical Relationships (Correlation and Regression)

* Statistical relationships batate hain ki do variables ke beech relation hai ya nahi — aur kitna strong hai

* Correlation: Sirf relation ka direction aur strength (−1 to +1).

* Regression: Ek variable ko use karke dusre ko predict karna.

* Why in AI/ML:

    - Feature selection, dependency understanding aur model building me use hota hai.

* Example:

    - Study hours vs Marks correlation.



```python
from scipy.stats import pearsonr

x = np.random.randint(1 , 10 , 10)
y = x * 5 + np.random.randint(-2 , 2 , 10)

corr , _ = pearsonr(x , y)
print("Correlation:", corr)
```

## Bayesian Statistics

* Bayesian statistics probability ko belief ke roop me dekhta hai, jo new data ke sath update hota hai.

* Formula:

    - P(H | D) = [P(D | H) * P(H)] / P(D)

* Example:

    - Agar kisi email ke spam hone ka prior chance 30% hai, aur usme spam word mile to belief update ho jata hai.

* Why in AI/ML:

    - Bayesian methods probabilistic reasoning, uncertainty handling aur model updating ke liye use hote hain.

    - Example: Naive Bayes Classifier, Bayesian Neural Networks.



```python
p_spam = 0.3
p_word_given_spam = 0.8
p_word = 0.5
posterior = (p_word_given_spam * p_spam) / p_word
print("Updated (Posterior) Probability:", posterior)
```

## Information Theory

* Information theory measure karta hai data ke andar “information” aur “uncertainty” ko.

* Key Concepts:

    - Entropy (H): Uncertainty measure

    - H(X) = −Σ P(x) log₂ P(x)

    - Information Gain: Kitni uncertainty kam hui data milne ke baad

* Why in AI/ML:

    - Decision Trees, Deep Learning aur Feature Selection me entropy aur information gain use hota hai.

* Example:

    - Decision Tree split choose karta hai jahan Information Gain sabse zyada hota hai.

```python
from math import log2

p = [0.4, 0.6]
entropy = -sum([pi * log2(pi) for pi in p])
print("Entropy:", entropy)
```
